```python
import pandas as pd
from textblob import TextBlob

data = {
    'text': [
        "I absolutely love this product, it works great!",
        "This is the worst experience I have ever had.",
        "It's okay, nothing special but does the job.",
        "Amazing quality and fast delivery, highly recommend!",
        "Terrible customer service, very disappointed.",
        "The product is fine, average performance.",
        "Best purchase I've made this year!",
        "Completely useless, broke after one day."
    ]
}

df = pd.DataFrame(data)
df.head()
```

```python
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['polarity'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['sentiment'] = df['text'].apply(get_sentiment)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='sentiment', palette='Set2', order=['Positive','Neutral','Negative'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```
